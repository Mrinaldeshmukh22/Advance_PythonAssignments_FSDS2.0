```python
Q1. If you have any, what are your choices for increasing the comparison between different figures on
the same graph?
```


One common technique for increasing the comparison between different figures on the same graph is to adjust the y-axis scale to zoom in on the relevant range of values. This can be achieved in Python using the ylim() function from the matplotlib library.

we use the set_ylim() function to set the y-axis limits to the range of 0 to 6, which helps to increase the comparison between the two lines plotted on the same graph.


```python
import matplotlib.pyplot as plt

# generate some example data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

# create a new figure and axis
fig, ax = plt.subplots()

# plot the data
ax.plot(x, y1, label='Line 1')
ax.plot(x, y2, label='Line 2')

# adjust the y-axis limits to focus on the range of interest
ax.set_ylim([0, 6])

# add a legend and axis labels
ax.legend()
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')

# show the plot
plt.show()
```

```python
Q2. Can you explain the benefit of compound interest over a higher rate of interest that does not
compound after reading this chapter?
```

Compound interest is a type of interest that is calculated on the initial amount of the investment as well as on the accumulated interest. This means that the interest earned is added to the principal amount, and then the interest is calculated again on the new amount. This process continues for each compounding period, which can be monthly, quarterly, or annually.

On the other hand, a higher rate of interest that does not compound only earns interest on the initial principal amount, and does not earn interest on the accumulated interest. This is known as simple interest.

The benefit of compound interest over a higher rate of interest that does not compound is that it allows your investment to grow faster over time. The effect of compounding can be significant over the long-term, as the interest earned on the accumulated interest can quickly add up.

```python
principal = 1000
interest_rate = 0.05
years = 10

# Calculate compound interest
compound_interest = principal * (1 + interest_rate) ** years

# Calculate simple interest
simple_interest = principal * (1 + (interest_rate * years))

print(f"Compound interest after {years} years: ${compound_interest:.2f}")
print(f"Simple interest after {years} years: ${simple_interest:.2f}")
```

```python
Q3. What is a histogram, exactly? Name a numpy method for creating such a graph.
```


A histogram is a graphical representation of the distribution of a dataset. It is a bar graph-like representation where the x-axis represents the range of values, and the y-axis represents the frequency of occurrence of values within that range.

In Python, NumPy provides a convenient method to create a histogram using the hist() function. The hist() function takes an array of data and a number of bins as input and returns a tuple containing the histogram values and bin edges.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a random dataset
data = np.random.randn(1000)

# Create a histogram with 10 bins
hist, bins = np.histogram(data, bins=10)

# Plot the histogram
plt.hist(data, bins=10)
plt.show()
```

```python
Q4. If necessary, how do you change the aspect ratios between the X and Y axes?
```

To change the aspect ratios between the X and Y axes in Python, you can use the set_aspect method of a Matplotlib axes object. The set_aspect method takes a single argument, which is the desired aspect ratio.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# plot some data on the axes object
ax.plot([1, 2, 3], [4, 5, 6])

# set the aspect ratio to be 1:1
ax.set_aspect('equal')

plt.show()
```

```python
Q5. Compare and contrast the three types of array multiplication between two numpy arrays: dot
product, outer product, and regular multiplication of two numpy arrays.
```

1) Dot Product:
The dot product of two arrays is calculated using the dot function. It is also known as matrix multiplication or inner product. This operation multiplies the corresponding elements of the arrays and then sums the results. The resulting array has the shape of the remaining dimensions after the common dimensions are removed.

```python
import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
c = np.dot(a, b)

print(c)
```

2) Outer Product:
The outer product of two arrays is calculated using the outer function. This operation multiplies each element of the first array with each element of the second array, resulting in a new array with the shape equal to the product of the shapes of the two input arrays.

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.outer(a, b)

print(c)
```

3) Regular Multiplication:
Regular multiplication of two arrays is performed using the * operator. This operation multiplies the corresponding elements of the arrays. If the two arrays have the same shape, then the resulting array will also have the same shape.

```python
import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
c = a * b

print(c)
```

The pmt function takes three arguments: the interest rate per period, the total number of payment periods, and the present value of the loan.

```python
import numpy as np
import numpy_financial as npf

#What is the monthly payment needed to pay off a $200,000 loan in 15 years at an annual interest rate of 7.5%?


interest_rate = 0.075 / 12  # monthly interest rate
total_periods = 15 * 12  # 15 year loan with 12 payments per year
loan_amount = 200000  # the amount of the loan

monthly_payment = npf.pmt(interest_rate, total_periods,loan_amount)

print("Monthly mortgage payment:", round(monthly_payment, 2))
```
